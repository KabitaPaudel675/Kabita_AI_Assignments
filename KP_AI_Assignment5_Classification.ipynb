```python
# import required modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# About the datset
This datasets is related to red variants of the Portuguese "Vinho Verde" wine. The refrence for the dataset is: [Cortez et al., 2009]. This dataset is also available from the UCI machine learning repository, https://archive.ics.uci.edu/ml/datasets/wine+quality.

Input variables (based on physicochemical tests):
1 - fixed acidity
2 - volatile acidity
3 - citric acid
4 - residual sugar
5 - chlorides
6 - free sulfur dioxide
7 - total sulfur dioxide
8 - density
9 - pH
10 - sulphates
11 - alcohol
Output variable (based on sensory data):
12 - quality (score between 0 and 10)

For this project we will set an arbitrary cutoff for the dependent variable (wine quality) at  6 or higher getting classified as 'good/1' and the remainder as 'not good/0'.

Relevant publication

P. Cortez, A. Cerdeira, F. Almeida, T. Matos and J. Reis. Modeling wine preferences by data mining from physicochemical properties. In Decision Support Systems, Elsevier, 47(4):547-553, 2009.

```python
# load dataset
df = pd.read_csv("https://raw.githubusercontent.com/KabitaPaudel675/AI_Assignment-1/main/winequality-red.csv")
df.head()
```

```python
## Replace values of columns by using DataFrame.loc[] property.
df.loc[df['quality'] <=5 , 'quality'] = 0
```

```python
## Replace values of columns by using DataFrame.loc[] property.
df.loc[df['quality'] >= 6, 'quality'] = 1
```

For classification using scikitlearn

```python
#splitting dataset into features and target variable
feature_cols = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide', 'pH', 'sulphates', 'alcohol']
X = df[feature_cols] # Features
y = df.quality # Target variable
```

```python
X
```

```python
y
```

```python
# split feature and target variables into training and testing sets (Testing set is 25% )
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression()

# fit the model with data
logreg.fit(X_train,y_train)
```

```python
#predicting the output for out test set
y_pred=logreg.predict(X_test)
y_pred
```

Classification Using GaussianNB

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
y_pred2=gnb.predict(X_test)
y_pred2
```

```python
df_out = X_test.copy()

df_out['LABEL'] = y_test
df_out['PREDICTED_LOGREG'] = y_pred
```

```python
df_out['PREDICTED_NB'] = y_pred2
```

Assessing model performance with scikitlearn

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
