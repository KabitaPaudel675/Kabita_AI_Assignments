# About the dataset
This dataset consists of corn and soybean yield of 12 states from Southeastern United States(Alabama,Arkansas, Florida, Georgia, Kentucky, Lousiana, Mississipi, North Carolina, South Carolina, Tennessee, Virginia, West Virginia) from 1980 and 2020 along with the average Standardised Precipitation-Evapotranspiration Index (SPEI) values for those years for each state. The SPEI is a multiscalar drought index based on climatic data.The SPEI is designed to take into account both precipitation and potential evapotranspiration (PET) in determining drought. 
In this project, we are trying to look at the trend of corn and soybean production and SPEI values in Kentucky from 1980 to 2020.
