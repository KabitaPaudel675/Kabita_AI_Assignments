For this exercise, we are doing paired t test comparing means of corn yield of Southeastern US in 1980 and 2020. Since we are comparing means of the same area in two different years, we are using paired t test for depoendent samples.
The  t-test works by checking the means from two samples to see if they are significantly different from each other. It does this by calculating the standard error in the difference between means, which can be interpreted to see how likely the difference is, if the two samples have the same mean (the null hypothesis).


```python

# Paired t-test for dependent samples

from scipy.stats import ttest_rel
```

```python
data1=df['1980']
```

```python
data2=df['2018']
```

```python
# compare samples
stat, p = ttest_rel(data1, data2)
print('Statistics=%.3f, p=%.3f' % (stat, p))
```

The test result shows the test statistic "t" is "-13.187" with p-value 0.000 that means the corn yield has decreased from 1980 to 2020 in the 12 Southeast US states and the decrement in the yield is highly statistically significant since p-value is less than 0.05.
