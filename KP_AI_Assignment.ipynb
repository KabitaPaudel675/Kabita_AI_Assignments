# About the dataset 
We are reading a comma delimited file (.csv) uploaded from local computer.
This file contains data related to approved permitted mine boundaries for surface and underground mines in Kentucky. Includes western and eastern coal fields, for active, inactive, and released permits. This data is derived from Permitted Mine Boundaries layer (a GIS file) found in Kentucky Geoportal website. This dataset is provided by Kentucky Division of Mine Permits.
The purpose of creating this dataset is to provide generalized boundaries of permitted coal mines, active and inactive, for the Commonwealth of Kentucky.
