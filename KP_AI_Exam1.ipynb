# 2. Is there statistical difference in the median_house_value of those with housing_median_age <=25 vs median_house_value of those with housing_median_age >25?
Answer:To find out whether there is statistical difference between median_house_value between these 2 groups: 1)Group 1: Housing_median_age<=25
2)Group 2: housing_median_age >25, we need to perform student t test. 

```python
## Creating two datasets for two different groups
df1=df[df['housing_median_age']<=25] ## for group 1: Housing_median_age<=25
df1
```

```python
df2=df[df['housing_median_age']>25] ##for group 2: Housing_median_age>25
df2
```

```python
data1=df1['median_house_value']
data2=df2['median_house_value']
print(data1)
print(data2)
```

```python
print(data1.mean())
```

```python
## For t-test for independent samples 
import scipy.stats as stats
# compare samples
stats.ttest_ind(data1, data2)
```

# 2 Answer: There is a statistically significant difference in the median_house_value of those with housing_median_age <=25 vs median_house_value of those with housing_median_age >25, t= -8.715, pvalue= 3.167821598965763e-18.
